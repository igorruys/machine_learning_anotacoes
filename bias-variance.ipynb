```python
import numpy as np
import scipy.stats as stat
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
np.random.seed(10)
```

```python
epsilon = np.random.normal(0, 5000, 50)
f = lambda x: 2000*x
X = np.random.normal(100, 20, 50)
Y = f(X) + epsilon
```

```python
N = 300
xx, yy = np.meshgrid(np.linspace(0, 200, 300), np.linspace(0, f(200), 300))
Z = stat.norm.pdf(xx, loc=100, scale=20) * stat.norm.pdf(yy, loc=f(xx), scale=5000)

fig, ax = plt.subplots(1)

#pcm = ax.pcolor(xx, yy, Z, cmap='Blues', shading='nearest')
#fig.colorbar(pcm, ax=ax, extend='max', label='linear scaling')
ax.contour(xx,yy, Z, cmap='Blues', levels=12)
ax.plot(X, Y, '.', color='tab:red')
ax.set_xlim([40, 160])
ax.set_ylim([f(40), f(160)])
ax.set_xlabel('Tamanho da casa ($m^2$)')
ax.set_ylabel('Pre√ßo (reais)')
```
